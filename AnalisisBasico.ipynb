```python
base_path = "./ConjuntosDatos/"
endutih_path = base_path + "ENDUTIH2020/conjuntos_de_datos/"
```

# INTERNET

## Estados

```python
estados_IDT = pd.DataFrame(endutih_res.groupby("ENT").apply(lambda x: 100*sum(x[x.P3_9_2==1].FAC_HOGAR)/sum(x.FAC_HOGAR)))
estados_IDT.columns = ["Internet"]
estados_IDT["NOMBRE"] = [nombre for nombre in estados.Nombre]
estados_IDT.loc[15,"NOMBRE"] = "MÃ©xico"
estados_IDT.loc[30,"NOMBRE"] = "Veracruz de Ignacio de la Llave"
estados_IDT
```

### Agregando Dominio Estrato

```python
estrato_dom_ent = pd.DataFrame(endutih_res.groupby(["DOMINIO","ESTRATO","ENT"]).apply(lambda x: 100*sum(x[x.P3_9_2==1].FAC_HOGAR)/sum(x.FAC_HOGAR)))
estrato_dom_ent.columns = ["Internet"]
estrato_dom_ent = estrato_dom_ent.reset_index()
estrato_dom_ent = pd.pivot_table(estrato_dom_ent, index=["DOMINIO", "ESTRATO"], columns=["ENT"], fill_value=0)
estrato_dom_ent.columns = estados.ENT
estrato_dom_ent
```

```python
estrato_dom_ent.to_csv(base_path+"estrato_dominio_estado_internet.csv")
```

## Grupos

```python
resumen = defaultdict(list)
columnas = ["Acceso", "Uso", "Aptitudes", "IDT_Mex"]
```

```python
print("ACCESO")
for grupo in sorted(endutih_completa.Grupo.unique()):
    acceso = IDTMex.componente_acceso(endutih_completa[endutih_completa.Grupo==grupo])
    resumen[f"Grupo {grupo}"].append(acceso)
resumen
```

```python
print("USO")
for grupo in sorted(endutih_completa.Grupo.unique()):
    uso = IDTMex.componente_uso(endutih_completa[endutih_completa.Grupo==grupo])
    resumen[f"Grupo {grupo}"].append(uso)
resumen
```

```python
print("APTITUD")
for grupo in sorted(endutih_completa.Grupo.unique()):
    aptitud = IDTMex.componente_aptitudes(endutih_completa[endutih_completa.Grupo==grupo])
    resumen[f"Grupo {grupo}"].append(aptitud)
resumen
```

```python
print("IDTMex")
for grupo in sorted(endutih_completa.Grupo.unique()):
    idtmex = IDTMex.IDTMex(endutih_completa[endutih_completa.Grupo==grupo])
    resumen[f"Grupo {grupo}"].append(idtmex)
resumen
```

```python
resumen_df = pd.DataFrame.from_dict(resumen, orient='index',
                       columns=columnas)
resumen_df.insert(0, column= "ESTRATO", value= [1]*1+[2]*4+[3]*5+[4]*5)
resumen_df
```

```python
resumen_df.groupby("ESTRATO").agg("mean")
```

```python
resumen_df.sort_values(by="IDT_Mex", ascending=False)
```
